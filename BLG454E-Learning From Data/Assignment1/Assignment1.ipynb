```python
#PART A: Examine training sets here. plots etc.
#Hint: df1_class0 = df1[df1['column_name']==0] creates a new dataframe and includes samples (rows) whose class values are 0
#hint: for data distribution plots use plt.scatter(x-axis,y-axis,other params. check online)
#Hint: for each new plot use plt.figure() so they won't overlap one to another
#-------
#to plot cov. matrices you may use: (find 'cov_mtrx' with yourself)
########       A2   #################
def visualizeCovarianceMatrix(dataSet):
    dataSetFeatures = dataSet.loc[:,['x1','x2']]
    covMatrix=np.cov(dataSetFeatures, rowvar=False)
    alpha = ['x1', 'x2']
    fig = plt.figure()
    ax = fig.add_subplot(111)
    cax = ax.matshow(covMatrix,interpolation='nearest') #cov. matrix of dataset1
    fig.colorbar(cax)
    ax.set_xticklabels(['']+alpha)
    ax.set_yticklabels(['']+alpha)
    print(covMatrix,"\n")
    return

    
class0_data1=df1[df1['y']==0]
print("Covariance Matrix of class 0 for training 1")
visualizeCovarianceMatrix(class0_data1)

class1_data1=df1[df1['y']==1]
print("Covariance Matrix of class 1 for training 1")
visualizeCovarianceMatrix(class1_data1)

class0_data2=df2[df2['y']==0]
print("Covariance Matrix of class 0 for training 2")
visualizeCovarianceMatrix(class0_data2)

class1_data2=df2[df2['y']==1]
print("Covariance Matrix of class 1 for training 2")
visualizeCovarianceMatrix(class1_data2)



###########################A.2 Comments#################################
#    For training1 data covariance matrix of both classes are equal. Feature x2 has a higher variance compared to feature x2.
#Covariance between features is negative and fairly low.
#
#    For training2 data covariance matrix of both classes are different. For class 0, feature x2 has a higher variance
#compared to feature x1. Covariance between features is positive. For class 1, feature x2 has a higer variance compared
#to feature x2 and the difference is greated compared to class 0. Covariance between features is negative.
#
#NOTE: Although the colors of the plots are similar, the scales are different as seen on the legend to the right of plots.   
#
```

```python
#########     B2    ############################

def trainBayes(trainingSet):
    class0 = trainingSet[trainingSet['y']==0]
    class0_mean = np.mean(class0.loc[:,['x1','x2']], axis=0)
    class0_cov = np.cov(class0.loc[:,['x1','x2']], rowvar=False)
    class1 = trainingSet[trainingSet['y']==1]
    class1_mean = np.mean(class1.loc[:,['x1','x2']], axis=0)
    class1_cov = np.cov(class1.loc[:,['x1','x2']], rowvar=False)
    idMatrix=True
    check0=class0.iloc[0,0]
    check1=class1.iloc[0,0]
    for i in range(class0_cov.shape[0]):
        for j in range(class0_cov.shape[0]):
            if(i==j):
                if(class0_cov[i][j]!=check0 or class1_cov[i][j]!=check1 ):
                    idMatrix=False
            else:
                if(class0_cov[i][j]!=0 or class1_cov[i][j]!=0):
                    idMatrix=False
            
    if (not np.allclose(class0_mean, class1_mean)) and np.allclose(class0_cov, class1_cov) :
        #print("Case1")
        def case1_g0(x):
            return (-1) * np.dot(np.dot(x-class0_mean,np.linalg.inv(class0_cov)),np.transpose(x-class0_mean))
        def case1_g1(x):
            return (-1) * np.dot(np.dot(x-class1_mean,np.linalg.inv(class1_cov)),np.transpose(x-class1_mean))
        return case1_g0, case1_g1 
    elif idMatrix :
        #print("Case2")
        def case2_g0(x):
            return (-1) * np.dot( (x-class0_mean),np.transpose(x-class0_mean) )
        def case2_g1(x):
            return (-1) * np.dot( (x-class0_mean),np.transpose(x-class0_mean) ) 
        return case2_g0, case2_g1
    else: 
        #print("Case3")
        def case3_g0(x):
            return (-0.5) * np.dot(np.dot(x-class0_mean,np.linalg.inv(class0_cov)),np.transpose(x-class0_mean)) + (-0.5) * np.log(np.linalg.det(class0_cov))
        def case3_g1(x):
            return (-0.5) * np.dot(np.dot(x-class1_mean,np.linalg.inv(class1_cov)),np.transpose(x-class1_mean)) + (-0.5) * np.log(np.linalg.det(class1_cov))
        return case3_g0, case3_g1
    
```

```python
##############      B3     ################
def testBayes(classifier0,classifier1,dataSet):
    testSet=dataSet.loc[:,['x1','x2']]
    labels= dataSet.loc[:,['y']] 
    predictedLabels=[]
    wrongPredictionCount=0;
    for i in range(testSet.shape[0]):
        if classifier0(testSet.iloc[[i],:]) > classifier1(testSet.iloc[i,:]):
            predictedLabels.append(0)
            if (labels.iloc[i]==1).bool() :
                wrongPredictionCount+=1
        else:
            predictedLabels.append(1)
            if(labels.iloc[i]==0).bool():
                wrongPredictionCount+=1
    return (wrongPredictionCount/labels.shape[0]*100), predictedLabels 
        
    
```

```python
(data1_classifier0, data1_classifier1)=trainBayes(df1)
test1 = pd.read_csv("test1.csv") 
(error, predictedLabels)=testBayes(data1_classifier0, data1_classifier1,test1  )
print("Error for Dataset1: ",error )
```

```python
(data2_classifier0, data2_classifier1)=trainBayes(df2)
test2 = pd.read_csv("test2.csv") 
(error, predictedLabels)=testBayes(data2_classifier0, data2_classifier1, test2 )
print("Error for Dataset2: ",error )
```

```python
##############      C1      ###############
data1= pd.concat([df1,test1])
```

```python
##############     C2 and C3   ##################
from sklearn.model_selection import KFold
kf = KFold(5,True,1)
data1_5FoldError=0
for trainIndex, testIndex in kf.split(data1):
   # print("Train: ",trainIndex," Test: ",testIndSex,"\n")
    (classifier0, classifier1) = trainBayes(data1.iloc[trainIndex,:])
    (error, predictedLabels) = testBayes(classifier0, classifier1, data1.iloc[testIndex,:])
    data1_5FoldError+=error
data1_5FoldError/=5
print("Dataset1 Error with 5-fold Cross-Validation: ",data1_5FoldError)
```

```python
data2= pd.concat([df2,test2])
data2_5FoldError=0
for trainIndex, testIndex in kf.split(data2):
   # print("Train: ",trainIndex," Test: ",testIndex,"\n")
    (classifier0, classifier1) = trainBayes(data2.iloc[trainIndex,:])
    (error, predictedLabels) = testBayes(classifier0, classifier1, data2.iloc[testIndex,:])
    data2_5FoldError+=error
data2_5FoldError/=5
print("Dataset2 Error with 5-fold Cross-Validation: ",data2_5FoldError)
```

```python
#############   C4   ##################
#Using 5-fold cross validation increased error for dataset 1 and decreased error for dataset2.
#K-fold cross validation strategies creates more reliable and generalizable models. 
#Using a random single split of dataset creates biased models due to randomization and may cause overfitting.
#This can be solved by using K-fold cross validation.
```
